```python
'''
Create file 'pupils.csv' and Name, Registered column.
Ask user to input name and registered or not registered.
write the user's input to the csv file
Allow the user to input until he decides to quit

Show the output of all the file content on the console.(read the file)
'''
```

```python
#Yemi
import csv

filename = 'pupils.csv'
with open(filename,'w') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Registered'])

def add_pupil():
    with open(filename, 'a') as file:
        writer = csv.writer(file)
        while True:
            name = input("Enter pupil's name (or type 'quit' to stop): ")
            if name.lower() == 'quit':
                break

            registered = input("Enter 'Registered' or 'Not Registered': ")
            writer.writerow([name, registered])
            print("Pupil added successfully.\n")
def display_pupils():
    print("\nCurrent pupils in the file:")
    with open(filename, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            print(row)

add_pupil()
display_pupils()
```

```python
#yemi
def read_file():
    while True:
        file_name = input("Please enter the file name: ")
        try:
            with open(file_name, 'r') as file:
                content = file.read()
                print("\nFile content:\n")
                print(content)
                break
        except FileNotFoundError:
            print("Error: The file '{}' does not exist. Please try again.".format(file_name))


read_file()
```
