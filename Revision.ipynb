```python
#Yemi
#Task 1
def find_name(names_list, name):
    if name not in names_list:
        return "Name not in list"
    else:
        position = names_list.index(name) + 1
        return name + " is number " + str(position) + " on the list"

names = ["Yemi", "Tee", "Kate", "Diana", "Simisola", "Matthew", "Deja"]

search_name = input("Enter a name to search for: ")
print(find_name(names, search_name))
```

```python
#Task 2

def cost_of_transport(km, mode): #function1
    if mode == "airplane":
        return 200 * km
    elif mode == "bus":
        return 20 * km

def cost_of_hotel(nights): #function2
    return 50 * nights

def cost_of_meals(days, type): #function3
    if type == "cafe":
        return 15 * 3 * days
    elif type == "fine dining":
        return 25 * 3 * days
def main():
    km = int(input("Enter the distance in km: "))
    mode = input("Enter the mode of transport (airplane or bus): ")
    nights = int(input("Enter the number of hotel nights: "))
    meal_type = input("Enter the type of meals (cafe or fine dining): ")

    transport_cost = cost_of_transport(km, mode)
    hotel_cost = cost_of_hotel(nights)
    meal_cost = cost_of_meals(nights, meal_type)

    if isinstance(transport_cost, int) and isinstance(meal_cost, int):
        total_cost = transport_cost + hotel_cost + meal_cost
        print("The total cost of your trip is " + str(total_cost) + " Dollars ")

main()
```

```python
#Task 3

def find_largest(numbers):
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest

def find_smallest(numbers):
    smallest = numbers[0]
    for num in numbers:
        if num < smallest:
            smallest = num
    return smallest

numbers = [10, 45, 23, 67, 5, 89, 32]
print("Largest number in the list:", find_largest(numbers))
print("Smallest number in the list:", find_smallest(numbers))
```

```python
Task 6

class PhoneBook:
    def __init__(self):
        self.contacts = {}

    def add_new_contact(self, name, phone_number):
        if name in self.contacts:
            print(name + " is already in your contacts.")
        else:
            self.contacts[name] = phone_number
            print(name + " has been added with phone number " + phone_number)

    def delete_contact(self, name):
        if name in self.contacts:
            del self.contacts[name]
            print(name + " has been deleted from your contacts.")
        else:
            print(name + " is not found in your contacts.")

    def show_phone_number(self, name):
        if name in self.contacts:
            return name + "'s phone number is " + self.contacts[name]
        else:
            return name + " is not found in your contacts."

    def display_all_contacts(self):
        if not self.contacts:
            print("Your contact list is empty.")
        else:
            print("Contact List:")
            for name, number in self.contacts.items():
                print("Name: " + name + ", Phone Number: " + number)


def main():
    phone_book = PhoneBook()

    while True:
        print("\nOptions:")
        print("1. Add New Contact")
        print("2. Delete Contact")
        print("3. Show Phone Number")
        print("4. Display All Contacts")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '1':
            name = input("Enter the contact name: ")
            phone_number = input("Enter the phone number: ")
            phone_book.add_new_contact(name, phone_number)
        elif choice == '2':
            name = input("Enter the contact name to delete: ")
            phone_book.delete_contact(name)
        elif choice == '3':
            name = input("Enter the contact name to view: ")
            print(phone_book.show_phone_number(name))
        elif choice == '4':
            phone_book.display_all_contacts()
        elif choice == '5':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

main()
```

```python
#Yemi
class Planet:
    def __init__(self, color, diameter):
        self.color = color
        self.diameter = diameter
        print("A new planet is born:", self.color, "and", self.diameter, "km in size")

    def getColor(self):
        return self.color

    def getDiameter(self):
        return self.diameter

    def getArea(self):
        radius = self.diameter / 2
        return 3.14 * (radius ** 2)

planetA = Planet('green', 150)
planetB = Planet('black', 100)

if planetA.getArea() > planetB.getArea():
    print(planetA.color, 'planet is bigger')
else:
    print(planetB.color, 'planet is bigger')
```

```python
#Yemi
class Planet:
    def __init__(self, color, size):
        self.color = color
        self.size = size

    def display_info(self):
        print(f"Planet is {self.color} and {self.size} cm wide")

class LifePlanet(Planet):
    def __init__(self, color, size, creatures):
        super().__init__(color, size)
        self.creatures = creatures

    def display_info(self):
        print(f"Planet is {self.color} and {self.size} cm wide with many creatures")

    def favorite_creature(self):
        print(f"Favorite creature is a {self.creatures[-1]}" if self.creatures else "No creatures found")

def main():
    planetA = Planet("blue", 25)
    planetB = LifePlanet("green", 20, ["Dragon", "Mermaid"])
    planetC = LifePlanet("purple", 50, ["Unicorn", "Phoenix", "Talking Teapot"])

    planetA.display_info()
    planetB.display_info(); print("B's", end=" "); planetB.favorite_creature()
    planetC.display_info(); print("C's", end=" "); planetC.favorite_creature()

main()
```
