```python
word='APPLES'
print('Guess the word with',len(word),'letters.')
print('A _ _ L _ _')
tries=3
correctletters=0
print('You have',tries,'guesses')
cg=''

while tries>0:
  guess=input('Guess one letter: ')
  guess=guess.upper()
  tries-=1 #tries=tries-1
  if guess in word:
    if guess=='A' or guess=='L':
      print('This letter is already given!')
      tries+=1
    elif guess in cg:
      print('You already guessed that letter!')
      tries+=1
    else:
      correctletters+=1
      cg=cg+guess
      print('Correct! You have',tries,'guesses left')
  else:
    print(guess,'is not in the word. You have',tries,'guesses left.')

print('You guessed',correctletters,'letters, The word was',word,'.')
```

```python
#YEMI
def validate_password(password):
    # Check if the password has more than 7 characters
    if len(password) <= 7:
        return "Password must be more than 7 characters."

    # Check if the password contains both letters and numbers
    if not any(char.isdigit() for char in password) or not any(char.isalpha() for char in password):
      print("Password must contain both letters and numbers.")
    # Check if the password contains both uppercase and lowercase letters
    if not any(char.isupper() for char in password) or not any(char.islower() for char in password):
      print("Password must contain both uppercase and lowercase letters.")

    # Check if the password contains any special characters
    if not password.isalnum():
      print("Password must not contain special characters.")

    # Check if the password starts with a number
    if password[0].isdigit():
      print("Password must not start with a number.")
      print("Password is valid.")

# Get the password from the user
user_password = input("Enter your password: ")

# Validate the password
result = validate_password(user_password)

# Print the result of the validation
print(result)


```
